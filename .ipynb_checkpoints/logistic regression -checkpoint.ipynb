```python
class logisticregression:
    def __init__(self,n_iter=100,learn_rate=0.01):
        self.learn_rate=learn_rate
        self.n_iter=n_iter
        self.weight=None
        self.bias=None
    def fit(self,x,y):
        #init parameters    
        n_samples,n_features=x.shape
        #initialize a array of weights for every features with zero;so now X1*w1,X2*w2~~~X1*0,X2*0
        self.weight=np.zeros(n_features)
        #initialize the bias b=0
        self.bias=0
        #gradient descend 
        #runing for loop n_iter times 
        for _ in range(self.n_iter):
            #formulating y=xw+b
            linearModel=np.dot(x,self.weight)+self.bias
            #converting to sigmoid limiting range between 0 and 1
            y_pred=self._sigmoid(linearModel)
            #update weights
            dw=(1/n_samples)*np.dot(x.T,(y_pred-y))
            db=(1/n_samples)*np.sum(y_pred-y)
            self.weight-=self.learn_rate*dw
            self.bias-=self.learn_rate*db
            
            
            
        
    def predict(self,x):
        #formulating y=xw+b
        linearModel=np.dot(x,self.weightight)+bias
        #converting to sigmoid limiting range between 0 and 1
        y_pred=self._sigmoid(linearModel)
        
        ypred_class=[1 if i>0.5 else 0 for i in y_pred]
        
        return ypred_class
        
    def _sigmoid(self,x):
        return 1/1+np.exp(-x)
            
        
        
```
